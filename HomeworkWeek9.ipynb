```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/ppp_excerpt.csv")
df
```

```python
## Run the code that resets the default display here to round the numbers, remove scientific notation, etc.
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df.info()
```

```python
## Run the code here to return all summary statistics in one go
df.describe()
```

```python
## What is the sum of all the forgiven loans
df["ForgivenessAmount"].sum()
```

```python
## What is the sum of all the forgiven loans
## format with dollar sign and thousands separators
## round to zero decimals
'${:,.0f}'.format(df["ForgivenessAmount"].sum())
```

```python
## store in variable called total_loan_forgiven
total_loan_forgiven = '${:,.0f}'.format(df["ForgivenessAmount"].sum())
```

```python
## checking it to make sure I wrote it correctly. It works.
total_loan_forgiven
```

```python
## return the median for CurrentApprovalAmount loans, InitialApprovalAmount loans and ForgivenessAmount loan amounts
df[["CurrentApprovalAmount", "InitialApprovalAmount", "ForgivenessAmount"]].median()
```

```python
##  Return the above as a frame
## I wasn't sure if by "as a frame" You meant in a easier to read display, as seen below,or in a sentence form
## It took me a while to figure out that I needed two sets of square brackets
df[["CurrentApprovalAmount", "InitialApprovalAmount", "ForgivenessAmount"]].median().to_frame("Median Amounts")
```

```python
## get the count, mean and median for CurrentApprovalAmount
df["CurrentApprovalAmount"].agg(["count", "mean", "median"])
```

```python
## I wanted to make this easier to read, so I did it again and added .to_frame
df["CurrentApprovalAmount"].agg(["count", "mean", "median"]).to_frame("Loan Aprovals")
```

```python
## automatically return a frame (without using to_frame()) for the count, mean and median for InitialApprovalAmount, CurrentApprovalAmount
df[['InitialApprovalAmount', 'CurrentApprovalAmount']].agg(['count', 'mean', 'median'])
```
